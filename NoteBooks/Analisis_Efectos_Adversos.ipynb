# Farmacovigilancia: Análisis de Efectos adversos en un ensayo clínico fase III

## Sobre el DataSet

los datos 'datos_estudio.csv' fueron tomados de hbiostat.org y son parte del repositorio de datos del Departamento de bioestadistica de la universidad de Vanderbilt. Provienen de un ensayo clínico fase III aleatorizado realizado por la farmacéutica GSK con un grupo 'Drug' y otro grupo 'Placebo'; Drug:Placebo de 2:1 (n = 1374 y 684) para un total de 2058 pacientes (https://hbiostat.org/data/)

Contiene el seguimiento de cada paciente de las 0-20 semanas (0,2,4,8,12,16,20); se registra una gran cantidad de parametros clínicos (Efectos adversos EA, Laboratorios clínicos, Signos vitales, ECG etc...).  Los efectos adversos a los que se les realizó el seguimiento fueron: Dolor de cabeza, nausea, diarrea, y exacerbación de la Enfermedad obstructiva cronica de vías aereas (coad), infección de tracto respiratorio superior, dispepsia y dolor abdominal

Realizaremos un análisis exploratorio, observaremos la distribución de algunas variables y ejecutaremos algunas pruebas de hipotesis utilizando librerias de python como pandas, seaborn, matplotlib y scipy.stats

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

df = pd.read_csv('datos_estudio.csv')
print (df.columns)

#Los efectos adversos tienen un valor numerico de 0-3 dependiendo de la gravedad
df[df['week'] == 8]['ab.pain'].value_counts()
```

## Análisis exploratorio y pruebas de hipótesis

Los datos de laboratorios clínicos tienden a presentar **skeweness** y algunos no siguen la distribución normal por lo que es conveniente utilizar la **mediana** para agregar información y **pruebas de hipótesis no paramétricas** como la de Mann-Whitney U. La librería de seaborn resulta bastante conveniente para gráficar información, se hará uso de funciones de distribución acumulada empiricas- **ecdf plots** para observar la distribución de los datos en la semana 8 de seguimiento.

```python
#Se trabajará en la semana 8; además,conviene categorizar los exámenes clínicos y signos vitales
df_4_week = df[df['week'] == 4]
df_8_week = df[df['week'] == 8]

electrolitos = ['potassium', 'sodium', 'chloride']
hematologicos = ['basophils.abs','eosinophils.abs','lymphocytes.abs','monocytes.abs','neutrophils','wbc', 'platelets','hematocrit']
hepaticos = ['bilirubin', 'alat', 'asat', 'ggt', 'alk.phos']
renales = ['bun', 'creatinine', 'uric.acid']
proteicos = ['albumin','protein']
vitales = ['sbp', 'hr', 'rr']
info = ['bmi', 'age']

```

### Boxen Plot

Se necesita crear un rango de edades para hacer visuzalizaciones estratificadas por edad, para esto se utilizara el método **cut()** de pandas; con esto podremos crear 5 rangos de edades para posteriormente hacer un boxenplot con seaborn. Los boxenplot son una mejor alternativa al clásico Boxplot para distribuciones de datos con Skewenees o con muchos outliers.

Hacemos la comparación entre 'sex' y 'trx'(tratamiento) y podemos graficar una linea de la mediana global para la semana 8 que facilite la comparación

seaborn facilita la manipulación y gráfica de datos utilizando DataFrames de pandas; los parametros **col** y **row** nos permiten crear subplots respecto a alguna variable categorica del DataFrame, en este caso 'sex' y 'trx'


```python
#Nuestros rangos de edad, son 5 rangos porque el parametro bins es = 5
labels = ['39-48','48-57','57-66','66-75','75-84'] 
df_8_week['rango_edades'] = pd.cut(df_8_week['age'], bins = 5, labels = labels)

#Se realiza el catplot, hacemos subplots ajustando el parametro col y row
g = sns.catplot(data = df_8_week, x = 'glucose', y = 'rango_edades', kind = 'boxen', col = 'trx', row = 'sex') 
g.refline(x = df_8_week['glucose'].median(), color = 'red') #se compara respecto a la mediana global
plt.show()
```

### Observando distribuciones: KdePlots, QQplots y pruebas de normalidad

Analicemos la variable 'age' y su distribución; nos damos cuenta que el kdeplot nos da una distribución que, a priori, no se asemeja a la campana gausseana normal.

¿Cómo podemos comprobar esto? Gráficamente podemos realizar un **QQplot**, que hace una relación entre los quantiles teóricos (de la distribución normal) y los observados; númericamente podemos utilizar la prueba de normalidad de **Shapiro-Wilk**, esto se puede realizar utilizando la libreria de Scipy.stats.

Además graficaremos otra serie de variables en las que se observan distribuciones que a priori no serían normales, prestele especial atención al conteo plaquetario. aquí el parametro **hue** nos permite hacer una comparación en el mismo gráfico de dos categorias Drug vs Placebo

```python
#Se realiza con kdeplot y observamos la diferencia en distribuciones por tratamiento con hue = 'trx'
sns.kdeplot(data = df_8_week, x = 'age', hue = 'trx')
plt.show()
```

```python
#Definimos un subset de variables que pueden presentar skeweness y distribuciones no normales
variables = ['sbp', 'bmi', 'glucose', 'platelets']

#Hacemos un loop para poder hacer varios subpot en lo axes definidos ( 2 x 2)
fig, axes = plt.subplots(nrows = 2, ncols= 2, figsize = (14,10))
axes = axes.flatten() #Fundamental para convertir el array de ejes en 1D y permitir un loop
for i, lab  in enumerate(variables):
    sns.kdeplot(data = df_8_week, x = lab, hue = 'trx', ax = axes[i])

plt.show()
```

### Prueba de Shapiro Wilk

Esta prueba de hipotesis nos permite corroborar si una distribución empírica es o no normal; es importante saber interpretar la prueba, **la Ho estipula que las distribucion de los datos de 'age' es igual a la distribución normal** (por esto se importa norm de scipy.stats.distributions), si obtenemos un p-valor menor a un alfa tendremos suficiente evidencia para rechazar esto y aceptar la hipotesis alternativa de que la distribución de 'age' es diferente de la normal

```python
#Importamos librerias necesarias para hacer las pruebas de hipotesis y verificar normalidad
from statsmodels.graphics.gofplots import qqplot 
from scipy.stats.distributions import norm
from scipy.stats import shapiro

# Realizamos el qqplot; dist = norm indica que queremos comparar respecto a la distribución normal
qqplot(df_8_week['age'], line = 's', dist = norm) 
plt.show()

#Aquí realizamos la prueba de Shapiro-Wilk y comparamos respecto a un alfa de 0.05
alpha = 0.05
stat, p = shapiro(df_8_week['age']) #nos interesa el p-valor

if p < alpha:
    print(f'El p valor para la prueba de Shapiro-Wilk fue de {p:.2e} < {alpha}')
    print('Conclusión: Rechazamos H0 - Los datos NO siguen distribución normal')
else:
    print(f'El p valor para la prueba de Shapiro-Wilk fue de {p:.2e} >= {alpha}')
    print('Conclusión: No rechazamos H0- Los datos podrían seguir distribución normal')

```

### Prueba no paramétrica de Mann-Whitney 

Me permite hacer pruebas de hipotesis entre grupos independientes en datos que no se comportan normalmente, esta prueba utiliza la mediana para comparar; la hipotesis nula Ho refiere a que la mediana de ambos grupos es igual; en este caso el p valor de 0.64 es mayor al alfa de 0.05 por lo que no existe evidencia de que la edad de los grupos Drug y Placebo sea diferente

```python
from scipy.stats import mannwhitneyu

#Definimos nuestro alfa y los dos grupos Drug vs Placebo
age_alpha = 0.05
drug_age = df_8_week[df_8_week['trx'] == 'Drug']['age']
placebo_age = df_8_week[df_8_week['trx'] == 'Placebo']['age']

#Realizamos la prueba de hipotesis entre grupos de Mann U whitney
u_stat, u_pval = mannwhitneyu(drug_age, placebo_age)

if u_pval < age_alpha:
   print(f"""  p_valor del test de Mann-Whitney es: {u_pval:.2f} < {age_alpha}
   por lo que hay evidencia para rechazar la Ho de que la mediana de edad de ambos grupos es igual
   existe evidencia a favor de una diferencia en la edad de ambos grupos  """)
else:
    print(f"""    p_valor del test de Mann-Whitney es: {u_pval:.2f} >= {age_alpha}
    por lo que no hay evidencia para rechazar la Ho de que la mediana de edad de ambos grupos es igual  """)
```

### ECDF plot entre Drug y Placebo

Graficaremos funciones acumulativas de probabilidad empíricas para ver como se distribuyen algunas variables y las compararemos entre 'Drug' y 'Placebo'.

Para esto utilizaremosla función ecdfplot() de seaborn y con el parámetro hue = 'trx' lograremos compararn las graficas respecto al tratamiento

```python
# se utilizara el ecdfplot de seaborn para graficar distribuciones de grupo Drug vs Placebo

# 'Embellecemos' el gráfico con matplotlib.pyplot
fig, axes = plt.subplots(nrows = 4, ncols = 2, figsize = (14,10), sharey = True)
fig.suptitle('Distribución Acumulada Empírica - Variables Hematológicas', 
             fontsize=16, fontweight='bold', y=0.98)

axes = axes.flatten() # permite aplanar el array de axes a 1-D y facilitar el loop

for i, lab in enumerate(hematologicos):
    ax = axes[i]
    sns.ecdfplot(data = df_8_week, x = lab, ax = ax, hue = 'trx', linewidth = 1.5, alpha = 0.8)
    ax.text(0.65, 0.15, s = lab, 
        transform=ax.transAxes,  # Coordenadas relativas (0-1)
        fontsize=8, 
        fontweight='bold',
        ) 
    # Estilo de ejes
    ax.set_xlabel('')
    ax.set_ylabel('Proporción Acumulada', fontsize=10)
    ax.tick_params(axis='x', labelsize=9)
    ax.tick_params(axis='y', labelsize=9)
    ax.grid(True, alpha=0.3, linestyle='--')
    
    # Personalizar leyenda
    if i == 0:  # Solo en el primer subplot
        ax.legend(title='Tratamiento', title_fontsize=10, fontsize=9, 
         loc='lower right', framealpha=0.9, labels=['Placebo', 'Drug'])
    else:
        ax.get_legend().remove()
plt.tight_layout()
plt.subplots_adjust(top=0.95)
plt.show()
```


Observamos que **existe cierta diferencia entre la ECDF de plaquetas para ambos grupos**, por lo que primero comprobaremos normalidad y luego realizaremos una prueba de hipotesis para ver si hay diferencia entre los grupos


```python
#Haremos lo mismo que lo que hicimos con la variable 'age' 
# qqplot-Shapiro-MannwhitneyU
platelets = df_8_week['platelets'].dropna()
qqplot(platelets, line = 's', dist = norm)
plt.show()

alpha = 0.05
stat, p = shapiro(platelets)

if p < alpha:
    print(f'El p valor para la prueba de Shapiro-Wilk fue de {p:.2e} < {alpha}')
    print('Conclusión: Rechazamos H0 - Los datos NO siguen distribución normal')
else:
    print(f'El p valor para la prueba de Shapiro-Wilk fue de {p:.2e} >= {alpha}')
    print('Conclusión: No rechazamos H0- Los datos podrían seguir distribución normal')
```

```python

platelets_alpha = 0.05
drug_platelets = df_8_week[df_8_week['trx'] == 'Drug']['platelets'].dropna()
placebo_platelets = df_8_week[df_8_week['trx'] == 'Placebo']['platelets'].dropna()
print(f'   mediana de plaquetas de Drug: {drug_platelets.median():.2f} xe9 L - mediana de plaquetas de placebo: {placebo_platelets.median():.2f} xe9 L')
u_stat, u_pval = mannwhitneyu(drug_platelets, placebo_platelets)

if u_pval < platelets_alpha:
   print(f"""   p_valor del test de Mann-Whitney es: {u_pval:.2f} < {platelets_alpha}
   por lo que hay evidencia para rechazar la Ho de que la mediana de conteo plaqueatario de ambos grupos es igual
   existe evidencia a favor de una diferencia en conteo plaquetario de ambos grupos  """)
else:
    print(f"""    p_valor del test de Mann-Whitney es: {u_pval:.2f} >= {platelets_alpha}
    por lo que no hay evidencia para rechazar la Ho de que la mediana del conteo plaquetario de ambos grupos es igual  """)
```


El p valor resultó ser de 0.01 por lo que **existe evidencia estadísticamente significativa para afirmar que hay diferencia en el conteo plaquetario de ambos grupos a la semana 8**; el grupo Placebo presenta menor conteo plaquetario.


### Variables categóricas

Finalmente analizaremos los efectos adversos, **(df_8_week[df_8_week['trx'] == 'Drug'][ea] != 0).sum()**  me da el conteo de todas las observaciones que NO son 0, osea que refiere algún tipo de efecto adverso; acordemonos que el grupo 'drug' tiene casi el doble de participantes que placebo, por lo que **queremos analizar las proporciones, no los conteos**, para esto dividimos entre 1374 y 684. Se loopea en la lista de efectos_adversos y se utiliza la librería de matplotlib para embellecer el gráfico.

Si resulta confuso entender el desfase, simplemente es una forma de poner una barra al lado de otra en la misma figura



```python
# Definimos nuestros efectos adversos a graficar
efectos_adversos = ['headache', 'ab.pain', 'dyspepsia', 'upper.resp.infect', 'coad', 'nausea', 'diarrhea']

#conteos de proporciones de efectos adversos
conteos_drug = []
conteos_placebo = []

#Hacemos un loop en efectos_adversos atentos a contar la proporción, no el total
for ea in efectos_adversos:
    conteo_drug = (df_8_week[df_8_week['trx'] == 'Drug'][ea] != 0).sum()/1374 #PROPORCIÓN
    conteo_placebo = (df_8_week[df_8_week['trx'] == 'Placebo'][ea] != 0).sum()/684 #PROPORCIÓN
    conteos_drug.append(conteo_drug)
    conteos_placebo.append(conteo_placebo)

#Graficamos y ajustamos ejes
fig, ax = plt.subplots(figsize = (14,10))
x = np.arange(len(efectos_adversos))  # Posiciones de las etiquetas
desfase = 0.35  # Ancho de las barras

drug_bar = ax.bar(x - desfase/2, conteos_drug, desfase, label='Drug', color='blue', alpha=0.5)
placebo_bar = ax.bar(x + desfase/2, conteos_placebo, desfase, label='Placebo', color='red', alpha=0.5)

ax.set_xlabel('Efectos Adversos', fontsize=12)
ax.set_ylabel('Proporción de Pacientes', fontsize=12)
ax.set_title('Comparación de Efectos Adversos: Drug vs Placebo, semana 8', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(efectos_adversos, rotation=45, ha='right')
ax.legend()
ax.grid(True, alpha=0.3)

plt.show()
```

Observamos que en el grupo 'Drug' se da la mayor proporción de efectos adversos (algo totalmente esperado), **sin embargo la exacerbación de coad es más frecuente en el grupo placebo, algo contraintuitivo** y que puede estar relacionado tal vez a la naturaleza de los pacientes elegidos; se realiza una gráfica de la proporción de pacientes con COAD para ambos grupos a lo largo de las semanas de seguimiento utilizando seaborn

```python
def create_coad_barplot_sns(df):
    # Crear una columna de presencia/ausencia
    df['coad_presence'] = (df['coad'] > 0).astype(int)
    
    # Filtrar solo semanas 0 y 8
    df_filtered = df[df['week'].isin([0,2,4,8])]
    
    # Crear el gráfico
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_filtered, x='week', y='coad_presence', hue='trx')
    
    plt.title('Proporción de COAD por tratamiento y semana')
    plt.xlabel('Semana')
    plt.ylabel('Proporción de COAD')
    
    plt.show()

create_coad_barplot_sns(df) 
```

### Prueba Chi2 de independencia

Realizaremos una prueba de Chi2 para ver si existe dependencia entre la variable 'trx' (tratamiento) y el efecto adverso de coad, la gráfica anterior nos señala que parece haber cierta relación entre el grupo placebo y la presencia de este efecto adverso.

Se utilizará chi2_contingency de scipy.stats, esta función acepta como input una tabla de contingencia que se peude obtener con **crosstab()** de pandas; para simplificar el analisis reemplazaramoes los valores de 2 por 1 para que se tenga una clásica tabla de contigencia de 2x2 (0 = ausencia del efecto adversos y 1 = presencia)

```python
from scipy.stats import chi2_contingency
#Queremos simplificar y dejar como 0-ausencia, 1-presencia del efecto adverso, por eso se utiliza el método replace()
df_8_week['coad'] = df_8_week['coad'].replace({2:1})

#Tabla de contingencia que sirve como input de la funcion chi2_contingency
contingency = pd.crosstab(df_8_week['trx'], df_8_week['coad']) 
print('tabla de contingencia:')
print(contingency)

#Realizamos la prueba de chi2 y comparamos con un alfa de 0.05
chi2, p_value, dof, expected = chi2_contingency(contingency)
alpha = 0.05
if p_value < alpha:
   print(f'el p-valor de {p_value:.4f} < {alpha} sugiere relación del tratamiento con el efecto adverso de COAD')
else:
    print(f' el p-valor de {p_value:.4f} >= {alpha} sugiere que no hay evidencia suficiente que indique relación entre el tratamiento y el COAD')
```
